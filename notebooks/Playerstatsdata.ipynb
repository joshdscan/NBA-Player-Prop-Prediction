```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import playergamelog
import pandas as pd
```

```python
def get_player_stats(player_name, active_player_list):
    # Retrieve a list of all active players
    active_player_list=pd.DataFrame(players.get_active_players())

    # Convert player name to title case
    player_name = player_name.title()
    
    # Retrieve the player's ID
    player_id = active_player_list.loc[active_player_list['full_name'] == player_name, 'id'].values[0]
    
    # Grab player games
    player_stats = pd.DataFrame(playergamelog.PlayerGameLog(player_id=player_id).get_data_frames()[0])
    
    return player_stats

```

```python
player_name = input("Player Full Name (Ex. 'Kevin Durant')")
player_stats = get_player_stats(player_name, active_player_list)
player_stats
```
